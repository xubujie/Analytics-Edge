## Numerical Features

```python
df.RiderID = df.RiderID.astype('category')
df.Index = df.Index.astype('category')
```

```python
# add some new features
df['Speed^2'] = df.Speed ** 2
df['Speed^3'] = df.Speed ** 3
df['Min_Gradient'] = df.Max_Gradient - 2* df.Average_Gradient
df['Min_heart_rate'] = df.Max_heart_rate - 2*df.Average_heart_rate
df['Average_point'] = (df.Highest_point + df.Lowest_point)/2
```

```python
corr = df.select_dtypes(include=['float64', 'int64']).iloc[:, 1:].corr()
plt.figure(figsize=(12,12))
sns.heatmap(corr, vmax = 1, square=True)
```

```python
cor_dict = corr['Power'].to_dict()
del cor_dict['Power']
print("List the numerical features decendingly by their correlation with Cadence:\n")
for ele in sorted(cor_dict.items(), key = lambda x: -abs(x[1])):
    print("{0}: \t{1}".format(*ele))
```

```python
sns.regplot(x = 'Speed', y = 'Cadence', data = df, color = 'Orange')
```

```python
sns.regplot(x = 'Power', y = 'Cadence', data = df, color = 'Red')
```

# Categorical Features

```python
def plot_boxtplot(df, x, y):
    plt.figure(figsize=(12,6))
    sns.boxplot(x = x, y = y, data = df)
    xt = plt.xticks(rotation = 45)
```

```python
print(df.select_dtypes(include=['object','category']).columns.values)
```

```python
plot_boxtplot(df, x='RiderID', y='Power')
```

```python
sns.countplot('RiderID', data = df, hue = df['Power_Category'])
```

```python
df.Date = pd.to_datetime(df.Date)
df['Year'] = df.Date.apply(lambda date:date.year)
df['Month'] = df.Date.apply(lambda date:date.month)
df['Day'] = df.Date.apply(lambda date:date.day)
```

```python
plot_boxtplot(df, 'Year', 'Power')
```

```python
plot_boxtplot(df, 'Month', 'Power')
```

```python
plot_boxtplot(df, 'Day', 'Power')
```

```python
df.Time = pd.to_datetime(df.Time)
df['Hour'] = df.Time.apply(lambda time: time.hour)
df['Min'] = df.Time.apply(lambda time:time.min)
```

```python
plot_boxtplot(df, 'Hour', 'Power')
```
